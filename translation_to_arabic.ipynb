### load libraries

```python
from deep_translator import GoogleTranslator
import re
from nltk.tokenize import word_tokenize
import nltk
nltk.download('wordnet')
nltk.download('punkt')

```

## Reading Data

```python
DATA=[]
with open(FILEPATH,'r',encoding='utf_8') as txt:
    for line in txt:
        #print(line, end='') # the end='' argument removes extra linebreaks
        Documont=''.join(line)
        DATA.append(Documont)
```

### Show Data

```python
DATA[:10]
```

### translation English language to Arabic language

```python
arabic=[]
for line in DATA[:100]:
    translation=GoogleTranslator(source='en', target='ar').translate(line)
    arabic.append(translation)
```

## Cleaning Text

```python
Context=[]
for sen in range(0, len(arabic)):
    # Remove all the special characters
    document = re.sub(r'\W', ' ', str(arabic[:sen]))
    # Converting to Lowercase
    document = document.lower()
    # Removing prefixed 'n'
    document = re.sub(r'n\s+', '', document)
    # tokenization
    mytokens = word_tokenize(document)
    Context = ' '.join(mytokens )
```
